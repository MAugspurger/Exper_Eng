```python
import numpy as np
import pandas as pd

num_com = 301

t_array_com = np.linspace(0,10,num_com)

dc = 15.0
freq1 = 1.0; amp1 = 4.0; ph1 = 0.0
freq2 = 3.0; amp2 = 2.0; ph2 = np.pi/2
freq3 = 20.0; amp3 = 0.7; ph3 = 0.0

dc_val = t_array_com*0.0 + dc
val1 = amp1*np.sin(t_array_com*freq1+ph1)
val2 = amp2*np.sin(t_array_com*freq2+ph2)
val3 = amp3*np.sin(t_array_com*freq3+ph3)

val_com = 0.0
val_com = dc + val1 + val2 + val3

val_low_pass = dc+val1
val_high_pass = dc + val3

data = pd.Series(data=val_com, index=t_array_com)
data.plot(xlabel = 'time(s)',ylabel='amplitude', ylim = [10,22],
          figsize=[6.0,3.0]);

# Visualization of filtered data
#data_low = pd.Series(data=val_low_pass, index=t_array_com)
#data_high = pd.Series(data=val_high_pass, index = t_array_com)
#data_low.plot()
#data_high.plot()
```

```python
# Plotting a stochastic signal
num_sto = 101
t_array_sto = np.linspace(0,10,num_sto)

dc = 10.0
limits = 10.0

dc_val = t_array_sto*0.0 + dc
val_sto = np.random.rand(num_sto)*limits +dc

data = pd.Series(data=val_sto, index=t_array_sto)
data.plot(xlabel = 'time(s)',ylabel='amplitude', ylim = [10,22],
          figsize=[6.0,3.0]);
```

```python
# Choose which plot above to average by commenting out
t_array = t_array_sto
num = num_sto
val_add = val_sto
#t_array = t_array_com
#num = num_com
#val_add = val_sto

val_new = val_add
len = 15

# Shift the values in the array forward by one time step
# Save each of the new arrays in val_new
for i in range(len):
    val_add = np.insert(val_add,0,0)
    val_add = np.delete(val_add,num)
    if i == 0:
        val_new = np.append([val_new], [val_add],axis = 0)
    else:
        val_new = np.append(val_new, [val_add],axis = 0)

# Average all the values across the val_new arrays
val_ave = t_array*0.0
for j in range(len):
    val_ave = val_ave + val_new[j]
val_ave = val_ave/float(len)

# Plot
time_cut = int((len+1)/2)
data = pd.Series(data=val_ave[(time_cut*2):], index=t_array[time_cut:num-time_cut])
data.plot(xlabel = 'time(s)',ylabel='amplitude', ylim = [10,22],
          figsize=[6.0,3.0]);
```
